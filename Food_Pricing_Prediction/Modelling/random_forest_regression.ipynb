```python
data_copied['YEAR'] = data_copied['Date'].dt.year
data_copied['MONTH'] = data_copied['Date'].dt.month
data_copied['DAY'] = data_copied['Date'].dt.day
data_copied.drop(['Date'], axis = 'columns', inplace = True)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.linear_model import Lasso, Ridge, LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
train = data_copied[data_copied['YEAR'].isin([2019, 2020])]
```

```python
test = data_copied[data_copied['YEAR'].isin([2021])]
```

```python
def fix(x):
    try:
        value=float(x)
    except:
        try:
            number=x.split(' ')
            value= number[0]+'.'+number[1]
            value= float(value)
        except:
            value=0
    return value

train['Food_prices'] = train['Food_prices'].apply(fix)
test['Food_prices'] = test['Food_prices'].apply(fix)
```

```python
print(train.shape, test.shape)
train= train[train['Food_prices']!=0]
test= test[test['Food_prices']!=0]
print(train.shape, test.shape)
```

```python
train.shape, test.shape
```

```python
train = pd.get_dummies(train)
```

```python
test = pd.get_dummies(test)
```

```python
print(train.shape, test.shape)
train.dropna(inplace = True)
test.dropna(inplace= True)
print(train.shape, test.shape)
```

```python
train.head()
train.tail()
```

```python
X_train = train.drop(['Food_prices'], axis = 1)
y_train = train['Food_prices']
X_test= test.drop(['Food_prices'], axis = 1)
y_test= test['Food_prices']
```

```python
train.corr()['Food_prices'].sort_values()
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
predx= lin_model.predict(X_train)
pred= lin_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
R_model = Ridge()
R_model.fit(X_train, y_train)
predx= R_model.predict(X_train)
pred= R_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
L_model = Lasso()
L_model.fit(X_train, y_train)
predx= L_model.predict(X_train)
pred= L_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
RFR_model = RandomForestRegressor(random_state=42)
RFR_model.fit(X_train, y_train)
predx= RFR_model.predict(X_train)
pred= RFR_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
KNR_model = KNeighborsRegressor()
KNR_model.fit(X_train, y_train)
predx= KNR_model.predict(X_train)
pred= KNR_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
SVR_model = SVR()
SVR_model.fit(X_train, y_train)
predx= SVR_model.predict(X_train)
pred= SVR_model.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(np.sqrt(mean_squared_error(y_test, pred)))
print(r2_score(y_train, predx))
print(r2_score(y_test, pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params= dict()
params['n_estimators']=[10,50,100,200]
params['max_depth']= [2,3,4,5,6,7,8,9,10]
params['bootstrap']= [True, False]
params['max_features']= ['auto', 'sqrt']
params['min_samples_split']= [2, 5, 10]
params['min_samples_leaf']= [1, 2, 5]

clf= RandomForestRegressor()
random= RandomizedSearchCV(estimator= clf,verbose=5, cv=5, n_iter=100, scoring='neg_root_mean_squared_error', n_jobs=-1,param_distributions=params)
```

```python
result= random.fit(X_train, y_train)
print(result.best_score_)
print(result.best_params_)
```
