```python
import pandas as pd
```

```python
def ingest_data(data_path: str, sheet_names: list = None) -> dict:
    if sheet_names is None:
        sheet_names = ['NATIONAL']
    data = pd.read_excel(data_path, sheet_names)
    return data
```

```python
sheet_names = ['ABIA', 'ABUJA', 'ANAMBRA', 'EBONYI', 'ENUGU', 'IMO', 
    'AKWA IBOM', 'BAYELSA', 'CROSS RIVER', 'DELTA', 'RIVERS', 'EDO',
    'ADAMAWA', 'BAUCHI', 'BORNO', 'GOMBE', 'BENUE', 'TARABA', 'YOBE', 
    'KOGI', 'KWARA', 'NASSARAWA', 'NIGER', 'PLATEAU', 'EKITI', 'LAGOS', 
    'ONDO', 'OGUN_', 'OSUN', 'OYO', 'JIGAWA_', 'KADUNA_', 'KANO_', 
    'KATSINA', 'KEBBI_', 'ZAMFARA_', 'SOKOTO']
```

```python
dict_data = ingest_data("SELECTED FOOD (JAN_2019 - NOV 2021).csv", sheet_names)
```

```python
def get_sheet_data(data: dict, sheet_name: str):
    return data[sheet_name]
```

```python
df_abia = get_sheet_data(dict_data, 'ABIA')
```

```python
def clean(df):
    # Transpose
    df.insert(0, "Item & Size", df_abia['ItemLabels'].astype(str) + ' ' + df['Unit of Measurement'].astype(str), True)
    df_abia.drop(['ItemLabels', 'Unit of Measurement'], axis='columns', inplace=True)
    transposed_df = df.T
    return transposed_df
```

```python
clean(df_abia)
```

```python
def clean_it(df):
    # Transpose
    df.insert(0, "Item & Size", df_abia['ItemLabels'].astype(str) + ' ' + df['Unit of Measurement'].astype(str), True)
    df_abia.drop(['ItemLabels', 'Unit of Measurement'], axis='columns', inplace=True)
    transposed_df = df.set_index("Item & Size").T
    return transposed_df
```
